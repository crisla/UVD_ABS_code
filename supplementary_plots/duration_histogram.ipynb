```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
import numpy as np
import pandas as pd
import csv
%matplotlib inline
```

```python
with open("../data/LFS_dur.csv", 'rb') as f:
    reader = csv.reader(f)
    LFS = list(reader)

spell1 = np.empty(0)
weight = np.empty(0)
for row in LFS[1:]:
    if row[0]== '' or row[1] == '':
        pass
    else:
        spell1 = np.append(int(row[0]),spell1)
        weight = np.append(int(row[0]),weight)
Nobs1 = len(spell1)
```

## Only the gaps

```python
with open("hist_gaps_1.csv", 'rb') as f:
    reader = csv.reader(f)
    gaps1 = list(reader)

c1_spell = np.empty(0)
c1_spell05 = np.empty(0)
c1plus_spell = np.empty(0)
c1plus_spell05 = np.empty(0)
```

```python
with open("hist_gaps_2.csv", 'rb') as f:
    reader = csv.reader(f)
    gaps2 = list(reader)

c2_spell = np.empty(0)
c2_spell05 = np.empty(0)
```

```python
for row in gaps1[1:200000]:
    if row[0]== '' or row[1] == '':
        pass
    elif row[0] == '1':
        c1_spell = np.append(int(row[5]),c1_spell)
        if row[1] == '1':
            c1_spell05 = np.append(int(row[5]),c1_spell05)
    else:
        c1plus_spell = np.append(int(row[5]),c1plus_spell)
        if row[1] == '1':
            c1plus_spell05 = np.append(int(row[5]),c1plus_spell05)
```

```python
for row in gaps2[1:200000]:
    if row[0]== '' or row[1] == '':
        pass
    else:
        c2_spell = np.append(int(row[4]),c2_spell)
        if row[0] == '1':
            c2_spell05 = np.append(int(row[4]),c2_spell05)
```

```python
histc1, count1 = np.histogram(c1_spell/30.0,bins=np.arange(0,48,0.5),range=(0,48),density=True)
histc1p, count2 = np.histogram(c1plus_spell/30.0,bins=np.arange(0,48,0.5),range=(0,48),density=True)
histc2, count3 = np.histogram(c2_spell/30.0,bins=np.arange(0,48,0.5),range=(0,48),density=True)
```

```python
histc105, count1 = np.histogram(c1_spell05/30.0,bins=np.arange(0,48,0.5),range=(0,48),density=True)
histc1p05, count2 = np.histogram(c1plus_spell05/30.0,bins=np.arange(0,48,0.5),range=(0,48),density=True)
histc205, count3 = np.histogram(c2_spell05/30.0,bins=np.arange(0,48,0.5),range=(0,48),density=True)
```

```python
xs = np.arange(0.5,48,0.5)
plt.figure(figsize=(28,10))
plt.plot(xs,histc1, lw=4, ls ="-",label="Correction 1 Gaps")
plt.plot(xs,histc1p, lw=4, ls ="-",label="Correction 1+ Gaps")
#plt.plot(xs,histc2, lw=4, label="Correction 2 Gaps")
#plt.xlim(0,48)
plt.xticks(range(0,48),fontsize=18)
plt.yticks(fontsize=24)
#plt.ylabel("Density", fontsize=32,labelpad=20)
plt.xlabel("Months", fontsize=32,labelpad=24)
plt.legend(loc="best", frameon=True, fontsize=32)
#plt.savefig('kgl6.pdf', format='pdf')
plt.show()
```
